```python
#Import the files
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#What kind of data is it
df = pd.read_csv('Dd5e_monsters.csv')
df.info()
```

All of this data is labeled as an object; for the analysis, we will have to convert some comulmns to a different data type. This will happen when we clean the data.

```python
# Cleaning and converting the data

df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

# Convert data types
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')
df['Race + alignment'] = df['Race + alignment'].astype('category')
df['HP'] = pd.to_numeric(df['HP'], errors='coerce')
df['Armor'] = df['Armor'].astype('category')
df['Speed'] = df['Speed'].astype('category')
df['Challenge rating  (XP)'] = pd.to_numeric(df['Challenge rating  (XP)'], errors='coerce')


# Save the cleaned data
df.to_csv('cleaned_data.csv', index=False)
#Seeing what the new data type is
df.info()
```

```python
plt.bar(df['Size'], df['Challenge rating  (XP)'])
plt.xlabel('Size of Species')
plt.ylabel('XP value')
plt.title('Size vs. Challenge Rating')
plt.show()
```

```python
plt.scatter(df['Size'], df['Challenge rating  (XP)'])
plt.xlabel('Size of Species')
plt.ylabel('XP value')
plt.title('Size vs. Challenge Rating')
plt.show()
```
